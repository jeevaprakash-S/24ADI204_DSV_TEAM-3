```python
print("TEAM 3 - WEEk 3")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv(
    "house_prices.csv",
    engine="python",
    on_bad_lines="skip"
)
df.dropna(axis=1, how='all', inplace=True)
print(df.info())
print(df.isnull().sum())

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print(df.isnull().sum())

for col in num_cols:
    if col == 'Index':
        continue
    plt.figure()
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()

Q1 = df['Price (in rupees)'].quantile(0.25)
Q3 = df['Price (in rupees)'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['Price (in rupees)'] >= lower_bound) & (df['Price (in rupees)'] <= upper_bound)].copy()

print(f"Original number of entries: {len(df)}")
print(f"Number of entries after removing 'Price (in rupees)' outliers (IQR method): {len(df_cleaned)}")

print(df_cleaned.info())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")

df = pd.read_csv("house_prices.csv", engine="python", on_bad_lines="skip")

print("First 5 Rows:")
print(df.head())

print("\nDataset Shape:", df.shape)

df.drop(columns=['Dimensions', 'Plot Area'], inplace=True, errors='ignore')
print("Removed 'Dimensions' and 'Plot Area' columns as they were entirely null.")
print("Updated Dataset Shape:", df.shape)

print("\nDataset Info:")
df.info()

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values Before Imputation:")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap Before Imputation")
plt.show()

print("\nHandling Missing Values...")

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in num_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values handled.")

print("\nMissing Values After Imputation:")
missing_values_after = df.isnull().sum()
print(missing_values_after[missing_values_after > 0])

plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap After Imputation")
plt.show()

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in num_cols:
    if col == 'Index':
        continue
    if not df[col].isnull().all():
        fig, ax = plt.subplots(1, 2, figsize=(12,4))

        sns.histplot(df[col], kde=True, ax=ax[0])
        ax[0].set_title(f"Distribution of {col}")

        sns.boxplot(x=df[col], ax=ax[1])
        ax[1].set_title(f"Boxplot of {col}")

        plt.show()

if 'location' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.countplot(x='location', data=df, order=df['location'].value_counts().nlargest(15).index, palette='viridis')
    plt.title("Count Plot of Top 15 Locations")
    plt.xlabel("Location")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

filtered_cat_cols = [col for col in cat_cols if col != 'location']

for col in filtered_cat_cols:
    if df[col].nunique() <= 15:
        plt.figure(figsize=(8,4))
        sns.countplot(x=df[col], order=df[col].value_counts().index[:10])
        plt.xticks(rotation=45)
        plt.title(f"Count Plot of {col}")
        plt.show()
    else:
        print(f"Skipping '{col}' (Too many unique values: {df[col].nunique()})")

print("\nSkewness:")
if not num_cols.empty:
    print(df[num_cols].skew())
else:
    print("No numerical columns to calculate skewness.")

print("\nEDA Deep Dive Completed Successfully!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Bivariate Analysis: Categorical Features vs. Price (in rupees) ---")

categorical_features_for_bivariate = [
    'location', 'Status', 'Transaction', 'Furnishing', 'facing', 'Bathroom', 'Balcony', 'Ownership'
]

for col in categorical_features_for_bivariate:
    if col in df.columns and 'Price (in rupees)' in df.columns:
        if df[col].nunique() > 10:
            top_categories = df[col].value_counts().nlargest(10).index
            df_filtered = df[df[col].isin(top_categories)]
            plot_title = f'Price (in rupees) Distribution by Top 10 {col}'
        else:
            df_filtered = df
            plot_title = f'Price (in rupees) Distribution by {col}'

        plt.figure(figsize=(12, 6))
        sns.boxplot(x=col, y='Price (in rupees)', data=df_filtered, palette='viridis')
        plt.title(plot_title)
        plt.xlabel(col)
        plt.ylabel('Price (in rupees)')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    elif col not in df.columns:
        print(f"Warning: Column '{col}' not found in DataFrame. Skipping.")
```
